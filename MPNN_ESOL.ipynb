```python
import numpy as np
np.random.seed(123)
import tensorflow as tf
tf.random.set_seed(123)
import deepchem as dc

# Load Delaney dataset
delaney_tasks, delaney_datasets, transformers = dc.molnet.load_delaney(
    featurizer='Weave', split='index')
train_dataset, valid_dataset, test_dataset = delaney_datasets
```

```python


metric = dc.metrics.Metric(dc.metrics.mean_squared_error, np.mean)
```

```python
n_atom_feat = 75
n_pair_feat = 14
# Batch size of models
batch_size = 64
```

```python
import numpy as np
np.random.seed(123)


model = dc.models.MPNNModel(
  len(delaney_tasks),
  n_atom_feat=n_atom_feat,
  n_pair_feat=n_pair_feat,
  T=3,
  M=5,
  batch_size=batch_size,
  learning_rate=.01,
  use_queue=False,
  mode="regression",
  edge_hidden_feats=64,
  num_step_message_passing=3)

# Fit trained model
model.fit(train_dataset, nb_epoch=100)
print("Evaluating model")
test_scores = model.evaluate(test_dataset, [metric], transformers)
print("Test MSE")
print(test_scores)
```

```python
print("Evaluating model")
test_scores = model.evaluate(valid_dataset, [metric], transformers)
print("Valid MSE")
print(test_scores)
```
